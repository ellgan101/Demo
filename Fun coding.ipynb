```python
def fun1(num, deno):
    output = ''
    if num * deno < 0:
        output += '-'
        num = abs(num)
        deno = abs(deno)
    output += str(num // deno)
    output += '.'
    ls = []
    rem = num % deno
    # separate the non-divisible case
    if rem != 0:
        while True:
            val = (rem * 10) // deno
            if val == 0:
                break
            if str(val) not in ls:
                ls.append(str(val))
                rem = (rem * 10) % deno
            else:
                bracket = ls.index(str(val))
                ls.insert(bracket, '(')
                ls.append(')')
                break
    else:
        ls.append(str(0))
    # Final presentation
    for i in ls:
        output += i
    return output

    
fun1(-1,-1)
```

```python
def prime(n):
    ls = []
    for num in range(2, n + 1):
        prime = True
        for i in range(2,num):
            if num % i == 0:
                prime = False
        if prime:
            ls.append(num)
    return ls

prime(10000)
```

```python
def helper(data, k):
    if k == 0:
         return 1
    s = len(data) - k
    if data[s] == '0':
        return 0
    # Result 
    result = helper(data, k - 1)
    if k >= 2 and int(data[s:s+2]) <= 26:
        result += helper(data, k - 2)
    return result

def num_ways(data):
    return helper(data, len(data))

num_ways('11111')
```

```python
def LCS(X, Y, i, j):
    if i<0 or j<0:
        return 0
    if X[i] == Y[j]:
        return 1 + LCS(X, Y, i-1, j-1)
    else:
        return max(LCS(X,Y,i,j-1), LCS(X,Y,i-1,j))

def counting(X,Y):
    return LCS(X,Y,len(X)-1,len(Y)-1)

counting('SHINCHAN','NOHARAAA')
```

```python
def LCS2(X,Y,m,n):
    memo = []
    for i in range(0,m):
        prev = 0
        for j in range(0,n):
            temp = memo[j]
            if X[i-1] == Y[j-1]:
                memo[j] = prev + 1
            else:
                memo[j] = max(memo[j],memo[j-1])
            prev = temp
    return memo[n]

LCS2('SHINCHAN','NOHARAAA',8,8)
```

```python
def ispalin(s):
    return s == s[::-1]

def palin(s):
    ans = -1
    length = len(s)
    for i in range(length):
        if s[i] != s[length-1-i]:
            if ispalin(s[i+1:length-1-i+1]):
                ans = i
            else:
                ans = length - 1 - i
            break
    return ans
            

palin('ifcldioxqllqxidlcfi')
```

```python
def rem(s):
    ls = []
    for i in range(len(s)):
        if s[i] not in ls:
            ls.append(s[i])
    return ls

def getValidIndex(s, j):
    i = 0
    while i < len(s):
        if s[i] <= j:
            i += 1
        else:
            break
    return i

def func(a,b,c):
    count = 0
    lena = sorted(rem(a))
    lenb = sorted(rem(b))
    lenc = sorted(rem(c))
    for k in lenb:
        counta = getValidIndex(lena,k)
        countc = getValidIndex(lenc,k)
        count += counta * countc
    return count

```

```python
def countTriplets(arr, r):
    ans = 0
    # Loop the j
    for j in range(1,len(arr)-1):
        if arr[j] // r in arr and arr[j] % r == 0:
            left = arr.count(arr[j] // r)
            right = arr.count(arr[j] * r)
            ans += left * right
    return ans

countTriplets([1] * 100,1)
```

```python
def dropEgg(n, k):
    # n: number of floors
    # k: number of eggs
    '''
    if n < 0 or k < 0:
        return 'Error'
    '''
    if n == 0 or k == 0:
        return 0
    elif n == 1:
        return 1
    elif k == 1:
        return n
    else:
        ans = 1 + max(dropEgg(n-1,k-1), dropEgg(1-1,k))
        for i in range(1, n+1):
            if ans > 1 + max(dropEgg(n-i,k-1), dropEgg(i-1,k)):
                ans = 1 + max(dropEgg(n-i,k-1), dropEgg(i-1,k))
    return ans

dropEgg(20,2) 
```

```python
def binomialCoeff(x, n, k):
 
    sum = 0;
    term = 1;
    i = 1;
    while(i <= n and sum < k): 
        term *= x - i + 1;
        term /= i;
        sum += term;
        i += 1;
    return sum;
 
# Do binary search to find minimum 
# number of trials in worst case.
def minTrials(n, k):
 
    # Initialize low and high as 
    # 1st and last floors
    low = 1;
    high = k;
 
    # Do binary search, for every 
    # mid, find sum of binomial 
    # coefficients and check if 
    # the sum is greater than k or not.
    while (low < high):
 
        mid = int((low + high) / 2);
        if (binomialCoeff(mid, n, k) < k):
            low = mid + 1;
        else:
            high = mid;
 
    return int(low);
 
# Driver Code
print(minTrials(2, 72))
```

```python
# Return the greatest power
def div(n,k):
    if n == 1 or n % k != 0 or k == 1:
        return 0
    else:
        if n % k == 0:
            return 1 + div(n // k, k)

def isugly(n):
    a = div(n,2)
    b = div(n,3)
    c = div(n,5)
    product = (2 ** a) * (3 ** b) * (5 ** c)
    if n // product == 1 and n % product == 0:
        return True
    else:
        return False

def ugly(n):
    count = 0
    i = 1
    while count < n:
        if isugly(i):
            count += 1
            i += 1
        else:
            i += 1
    return i-1

ugly(2)
```

```python
def getNthUglyNo(n): 
  
    ugly = [0] * (n+1) # To store ugly numbers 
  
    # 1 is the first ugly number 
    ugly[0] = 1
  
    # i2, i3, i5 will indicate indices for 2,3,5 respectively 
    i2 = i3 = i5 = 0
  
    # set initial multiple value 
    next_multiple_of_2 = 2
    next_multiple_of_3 = 3
    next_multiple_of_5 = 5
  
    # start loop to find value from ugly[1] to ugly[n] 
    for l in range(1, n+1): 
  
        # choose the min value of all available multiples 
        ugly[l] = min(next_multiple_of_2, next_multiple_of_3, next_multiple_of_5) 
  
        # increment the value of index accordingly 
        if ugly[l] == next_multiple_of_2: 
            i2 += 1
            next_multiple_of_2 = ugly[i2] * 2
  
        if ugly[l] == next_multiple_of_3: 
            i3 += 1
            next_multiple_of_3 = ugly[i3] * 3
  
        if ugly[l] == next_multiple_of_5:  
            i5 += 1
            next_multiple_of_5 = ugly[i5] * 5
  
    # return ugly[n] value 
    return ugly[-1], next_multiple_of_2, next_multiple_of_3, next_multiple_of_5, i2, i3, i5

getNthUglyNo(9)
```

```python
def fib(n):
    first = 1
    second = 1
    if n <= 2:
        return 1
    else:
        for i in range(2,n+1):
            nex = first + second
            first = second
            second = nex
    return nex

fib(400009)
```
